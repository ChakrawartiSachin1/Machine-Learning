# Simple Linear Regression

**Simple Linear Regression**

Simple linear regression is a statistical method used to model the relationship between a single independent variable \( X \) and a dependent variable \( Y \). It assumes that there is a linear relationship between the two variables, which can be represented by a straight line. The goal is to find the best-fitting line that minimizes the difference between the observed values of \( Y \) and the values predicted by the linear model.



```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

+ Create Sample Data


```python
# Create some sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape to column vector
y = np.array([2, 3, 4, 5, 6])
```

+ Create and Fit the Model

```python
# Create a LinearRegression object
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)
```

+ Print Learned Parameters

```python
# Print the learned parameters
print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])
```

+ Make Predictions

```python
# Make predictions
new_X = np.array([6, 7, 8]).reshape(-1, 1)  # Reshape to column vector
predictions = model.predict(new_X)
print("Predictions:", predictions)
```

+ Data Visualization

```python
# Plot the sample data
plt.scatter(X, y, color='blue', label='Sample Data')

# Plot the linear regression line
plt.plot(X, model.predict(X), color='red', label='Linear Regression Line')

# Add labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
